```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
# keeps plots inline for the notebook 
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
```

```python
x
```

```python
# numpy has functions which are inbuilt and can be used for mathematical purposes
sinx = np.sin(x)
sinx
```

```python
# it's easier to see this represented when we express it as a graph with matplot 
pp.plot(x,sinx)
```

```python
cosx = np.cos(x)
```

```python
pp.plot(x,sinx)
pp.plot(x,cosx)

pp.legend(['sin(x)','cos(x)'])
```

```python
# matplotlib has lots of interesting functions and is definitely worth checking out 
pp.plot(x,sinx, 'x')
pp.plot(x,cosx, 'o')
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2

pp.plot(x,y)
pp.plot(x,z)

pp.legend(['sin(x) * cos(x)', 'cos(x)^2 - sin(x)^2'])
```

```python
# .dot treats the two one dimensional arrays as vectors
np.dot(sinx,cosx)
```

```python
np.outer(sinx,cosx)
```

```python
# if you take a one dimensional element and add a number to it, it will add to every element in the array
v = np.linspace(0,10,5)
v + 1
```

```python
# if we add a 1D array to a 2D array it will add it to every row
vv = np.outer(v,v)
vv + v
```

```python
# if we want to add it to every column instead we can do so by transforming it into a column vector then adding
vv + v[:,np.newaxis]
```
