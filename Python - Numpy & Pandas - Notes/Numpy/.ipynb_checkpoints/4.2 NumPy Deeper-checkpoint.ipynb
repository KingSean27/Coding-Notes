```python
import numpy as np
```

```python
# we can use linspace to give a regularly spaced array, definte start, end and number of steps 

a = np.linspace(0,10,5, dtype=np.int64)
print(a)
```

```python
# we can use np.random to generate random arrays 

b = np.random.standard_normal((2,4))
print(b)
```

```python
# comparisons with arrays are also vectorised, so can be used on the whole array easily
# this output will be a a boolean array 

names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])

print(names)
print(names == 'Bob','\n')
print(a)
print(a < 3)
```

```python
# these boolean masks can be generated using artithemtic operators just like in base python 
# note that selecting data from an array by boolean indexing will always create a copy of the data 

mask = (names == 'Bob')|(names == 'Joe')
print(mask)
```

```python
# we can use these boolean arrays to set data for an array, 
# below example sets all values under 0 as 0 

b[b < 0] = 0 
print(b)
```

```python
# .transpose() can be used to switch columns and rows
# note we can use .T instead of transpose
# for higher dimensional arrays we use swapaxes and specify the axes we want to swap 

print(b.transpose())
```

```python
# the arange and reshape commands can be used to generate arrays of sizes we define
# we need to specify exactly otherwise we will get an error 

c = np.arange(32).reshape((8, 4))

print(c)
```

```python
# a ufunc is a universal function that performs operations on ndarrays

print(np.sqrt(c))
```

```python
# the power of NumPy comes from the fact that is allows us to write data processing tasks withoout loops 
# as we know this is called vectorisation and it is many orders of magnitude faster in terms of computing power
# than normal python calculations
```

```python
# we can save arrays as files then reload them into python 

np.save('example.npy',c)

d = np.load('example.npy')
print (d)
```

```python
# here we have generated two random arrays then stacked them onto each other vertically using the np.vstack
# we can also do this horizontally with np.hstack 

e = np.random.random((2,3))
f = np.random.random((2,3))

print('array e is:\n', e,'\n\narray f is:\n', f)

g = np.vstack([e,f])
print('\ncombined they are:\n', g)
```
