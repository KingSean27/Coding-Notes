```python
import pandas as pd 
import numpy as np
import yfinance as yf
import plotly.express as px
from IPython.display import Image
```

```python
stock_returns = yf.download(['MSFT', 'AAPL', 'LMT', 'BABA'], start="2014-10-22", end="2024-10-22")
stock_returns
```

```python
# Note how the data we have returned is a multi-index, we can access this using two levels on the index
stock_returns['Close']['LMT']
```

```python
fig = px.line(stock_returns['Close']['LMT'], color_discrete_sequence=["#7873BB"])
fig.update_yaxes(title="Share Price ($)", visible=True, showticklabels=True)
fig.show()
```

```python
percent_changes = (stock_returns['Close'].pct_change() + 1).cumprod()
percent_changes
```

```python
# Re-assign the resulting DataFrame to 'Percent' under a new level in the MultiIndex
# Add the 'Percent' level columns to the original DataFrame
percent_changes.columns = pd.MultiIndex.from_product([['Percent'], percent_changes .columns])
stock_returns = pd.concat([stock_returns, percent_changes ], axis=1)
```

```python
fig = px.line(stock_returns['Percent'], color_discrete_sequence=px.colors.qualitative.G10)
fig.update_yaxes(title="Percent Return", visible=True, showticklabels=True)
fig.show()
```

```python
# We can also assign portfolio weights to our portfolio, in this case we have evenly weighted the 4 stocks

weights = np.array([0.25, 0.25, 0.25, 0.25])
stock_returns['Portfolio Returns'] = stock_returns['Percent'].dot(weights)
stock_returns
```

```python
fig = px.line(stock_returns['Percent'], color_discrete_sequence=px.colors.qualitative.G10)
fig.add_scatter(y=stock_returns['Portfolio Returns'], x=stock_returns.index, mode='lines', 
                name='Portfolio', line=dict(color='purple'))
fig.update_yaxes(title="Percent Return", visible=True, showticklabels=True)
fig.show()
```

# Measuring Risk:

```python
# Variance of the returns is used to measure risk
# It is whether there is more or less spread around the mean

Image(filename="Variance.jpg")
```

```python
# Assets in a portfolio are correlated so measuring the portfolio of returns is not simply a case of multiplying/adding
# Need to introduce correlation factor too, for a two stock portfolio
# Last term is correlation times the two variances i.e. covariance

Image(filename="Portfolio Variance 2 Stocks.jpg")
```

```python
# This can be written in a smarter way in matrix notation

Image(filename="Matrix Notation.jpg")
```

```python
# In essence we need to calculate in python: Portfolio Variance = Weights transposed x (Covariance Matrix x Weights)
# Can do this using the covariance matrix

daily_returns = stock_returns['Close'].pct_change()
covariance = daily_returns.cov()
covariance
```

```python
# We need to annualize our volatility here as we are otherwise dealing with daily numbers 
# Multiple by 250 which is the number of trading days in the year

covariance_matrix = daily_returns.cov() * 250
covariance_matrix
```

```python
portfolio_variance = np.dot(weights.T, np.dot(covariance_matrix, weights))
print("Annual portfolio variance is ", (np.round(portfolio_variance, 3) * 100), '%')
```

```python
# We can take the square root for the stdev like this 
# Its more interpretable than variance because itâ€™s on the same scale as the returns themselves.

portfolio_stddev = np.sqrt(np.dot(weights.T, np.dot(covariance_matrix, weights)))
print("Annual standard dev is ", (np.round(portfolio_stddev, 3) * 100), '%')
print("The returns will be expected to fluctuate around the mean by ths much i.e.  Mean Â± 21.5%")
```
