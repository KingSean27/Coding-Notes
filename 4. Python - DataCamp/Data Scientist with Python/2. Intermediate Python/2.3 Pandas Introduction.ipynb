Pandas is one of the key data structures in python, you have encountered it before... Similar to an excel table, we have columns and rows. 

You can use dictionaries to make a pandas dataframe however that is not ideal. It is better to import from csv files. 

By default, the index is set as an integer, you have to specify the index if you want one. 

```
import pandas as pd
import numpy as np

stock_pitches = pd.read_csv("stocks_pitches.csv")

stock_pitches
```

```
stock_pitches = pd.read_csv("stocks_pitches.csv", index_col = 2)

stock_pitches
```

There are numerous ways we can select data in pandas. The first is to use square brackets to 

These individual columns are known as a series by combining the multiple series together we get a dataframe. 

```
stock_pitches["Pitch Price"]
```

```
print(type(stock_pitches["Pitch Price"]))

print(type(stock_pitches))
```

We can use two columns to return a dataframe object, this can be used to return two or more columns. 

To get the horizontal data, we use list slicing

```
stock_pitches[["Pitch Price", "Price %", "Total %"]]
```

```
stock_pitches[2:5]
```

Whilst using the square brackets is useful we can extend our funtionality better by use the loc and iloc function 

loc = label based 

iloc = integer based

To get the information back as a dataframe, we have to put it inside multiple sets of brackets. We can also use these to splice the dataframe up across rows and columns. We can also change the order of these too.

We can use : to include all things along that axis

```
stock_pitches.loc["BABA"]
```

```
stock_pitches.loc[["BABA"]]
```

```
stock_pitches.loc[["BABA", "FL", "GM"]]
```

```
stock_pitches.loc[["BABA", "FL", "GM"], ["Pitch Date", "Name", "Curr"]]
```

```
stock_pitches.loc[ :, ["Pitch Date", "Name", "Curr"]]
```

```
# iloc used for integers
stock_pitches.iloc[2:7, 2:6]
```

We can use boolean operators on pandas dataframes. 

We will return True or False values depending on how they compare to the operator

We can then use these to select values from the initial dataframe by using square brackets 

```
positive_return = stock_pitches["Total %"] > 0 

positive_return
```

```
stock_pitches[positive_return]
```

```
# we could do this process in one line instead if we wanted to 
stock_pitches[stock_pitches["Total %"] > 0]
```

We can use the numpy logical functions to further filter the data, this allows two or more conditions for out data. These work the same way they work for numpy in pandas dataframe

```
in_between = np.logical_and(stock_pitches["Total %"] > 0 , stock_pitches["Total %"] < 0.5)
in_between
```

```
stock_pitches[in_between]
```

```
# can also write in a single line 
stock_pitches[np.logical_and(stock_pitches["Total %"] > 0 , stock_pitches["Total %"] < 0.5)]
```
