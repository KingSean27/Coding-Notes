Dictionaries are one of the most common data types in python. It has key value pairs to list data. Covered this before in more detail. 

Dictionaries use curly brackets.

```
middle_east = {"Iraq":40.5, "Iran":82.3, "Saudi Arabia":20.1}

print(middle_east["Iran"])

middle_east["Israel"] = 12.5

print(middle_east)
```

```
"Israel" in middle_east
```

```
del(middle_east["Israel"])

print(middle_east)
```
