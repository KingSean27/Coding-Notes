```python
# strips date, name & price out of every .csv file in folder and merges into new text file
import os
directory_in_str = '/cath/homes2/zcbtsdl/Documents/Test Project/Historical Pricing'

directory = os.fsencode(directory_in_str)

outfile = open('NASDAQ INC. Pricing', 'wt')
storeoutput = ""

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    ###print (filename)
    if '.csv' in filename:
        ###print (filename)
        infile = open('{}'.format(filename),'rt')
        for line in infile: 
            if 'Product':        ### if 'Nasdaq' in line:
                line = line.replace(',', ' ')
                ###print (line)
                preplist = line.split()
                cleanname = infile.name[0:-4]
                cleanstock = str(preplist[:1]).strip('[]')
                cleanstock = cleanstock [1:-1]
                cleanprice = preplist [-1]
                cleanoutput = cleanname + " " + cleanstock + " " + cleanprice
                storeoutput = storeoutput + cleanoutput + "\n"
            else: 
                continue

                            
print (storeoutput)
```

```python
print(storeoutput, end = '', file = outfile)
outfile.close()
```

```python
# creates a dictionary containg the date and price of every entry for that stock 
infile = open('NASDAQ INC. Pricing','r')


NASDAQ = {}

for line in infile: 
    if 'NASDAQ' in line:
        prepstock = str.split(line)
        #print (prepstock)
        prepdate = str(prepstock[:1])
        prepdate = prepdate [2:-2]
        #print (prepdate)
        prepprice = str(prepstock[-1:])
        prepprice = prepprice [2:-2]
        #print (prepprice)
        NASDAQ[prepdate] = prepprice
        
        
print (NASDAQ)
    
print (NASDAQ.get("2018-11-29"))
```

```python
# cycles through a time vector and if the day is an entry in the dictionary puts it into the array in 
# chronological order

Datecycle = np.datetime64('2018-06-01') 
plotarray = []

for i in range (600):
    Datecycle = Datecycle + np.timedelta64(1,'D')
    Datestring = str(Datecycle)
    if NASDAQ.get(Datestring):
        print(NASDAQ.get(Datestring))
        plotarray.append(NASDAQ.get(Datestring))
        
print (plotarray)
```

```python
# plots array
plotarray = list(map(float, plotarray))

x = np.linspace(0, len(plotarray)-1, len(plotarray))
y = plotarray
  
pp.plot(x, y) 
pp.title('Share Price')
pp.legend(['NDAQ'])
```
